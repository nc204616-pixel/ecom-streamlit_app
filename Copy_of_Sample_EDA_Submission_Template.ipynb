## ***1. Know Your Data***

### Import Libraries

````python
# Import Libraries


# Data downloading from google
import gdown

# Data manipulation and analysis
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
# Load Dataset

# Download titles.csv
url_titles = "https://drive.google.com/uc?id=1IPTkKEOg-NH00fWgHKgFj6vK6ONZ8Y-K"
gdown.download(url_titles, "titles.csv", quiet=False)

# Download credits.csv
url_credits = "https://drive.google.com/uc?id=1B_ds97j4bLaOKoFaZrTdc47Q0AJrA9Qo"
gdown.download(url_credits, "credits.csv", quiet=False)

# Load datasets into pandas
titles = pd.read_csv("titles.csv", low_memory=False)
credits = pd.read_csv("credits.csv", low_memory=False)

print("‚úÖ Datasets Loaded Successfully!")
print("Titles shape:", titles.shape)
print("Credits shape:", credits.shape)
````

### Dataset First View

````python
# Dataset First Look

print(f"Titles dataset shape: {titles.shape}")
print(f"Credits dataset shape: {credits.shape}")
display(titles.head())
display(credits.head())
````

### Dataset Rows & Columns count

````python
# Dataset Rows & Columns count

# Titles dataset shape
titles_rows, titles_cols = titles.shape
print(f"Titles Dataset ‚Üí Rows: {titles_rows:,} | Columns: {titles_cols}")

# Credits dataset shape
credits_rows, credits_cols = credits.shape
print(f"Credits Dataset ‚Üí Rows: {credits_rows:,} | Columns: {credits_cols}")
````

### Dataset Information

````python
# Dataset Info

print("üìò TITLES DATASET INFO:\n")
titles.info()

print("\nüìó CREDITS DATASET INFO:\n")
credits.info()

# --- Quick summary statistics for numeric columns ---
print("\nüìä TITLES DATASET ‚Äî NUMERIC SUMMARY:")
display(titles.describe())

print("\nüìä CREDITS DATASET ‚Äî NUMERIC SUMMARY:")
display(credits.describe())

# --- Summary for categorical columns ---
print("\nüî† TITLES DATASET ‚Äî CATEGORICAL SUMMARY:")
display(titles.describe(include=['object']))

print("\nüî† CREDITS DATASET ‚Äî CATEGORICAL SUMMARY:")
display(credits.describe(include=['object']))
````

#### Duplicate Values

````python
# Dataset Duplicate Value Count

# --- Check duplicate rows in Titles dataset ---
titles_duplicates = titles.duplicated().sum()
print(f"Titles Dataset ‚Üí Duplicate Rows: {titles_duplicates}")

# --- Check duplicate rows in Credits dataset ---
credits_duplicates = credits.duplicated().sum()
print(f"Credits Dataset ‚Üí Duplicate Rows: {credits_duplicates}")

# --- (Optional) Display duplicate rows if any ---
if titles_duplicates > 0:
    print("\nüîç Sample duplicate rows in Titles Dataset:")
    display(titles[titles.duplicated()].head())

if credits_duplicates > 0:
    print("\nüîç Sample duplicate rows in Credits Dataset:")
    display(credits[credits.duplicated()].head())
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count

# --- Titles Dataset ---
print("üìò TITLES DATASET ‚Äî Missing Values:\n")
missing_titles = titles.isnull().sum()
print(missing_titles[missing_titles > 0])
print(f"\nTotal Missing Columns in Titles Dataset: {(missing_titles > 0).sum()}")

# --- Credits Dataset ---
print("\nüìó CREDITS DATASET ‚Äî Missing Values:\n")
missing_credits = credits.isnull().sum()
print(missing_credits[missing_credits > 0])
print(f"\nTotal Missing Columns in Credits Dataset: {(missing_credits > 0).sum()}")
````

````python
# Visualizing the missing values

# --- Titles Dataset ---
plt.figure(figsize=(10, 6))
sns.heatmap(titles.isnull(), cbar=False, cmap="coolwarm", yticklabels=False)
plt.title("üîç Missing Values Heatmap ‚Äî Titles Dataset")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()

# --- Credits Dataset ---
plt.figure(figsize=(10, 6))
sns.heatmap(credits.isnull(), cbar=False, cmap="coolwarm", yticklabels=False)
plt.title("üîç Missing Values Heatmap ‚Äî Credits Dataset")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
````

Dataset Description

This data set was created to list all shows available on Amazon Prime streaming, and analyze the data to find interesting facts. This dataset has data available in the United States.

This dataset has two files containing the titles (titles.csv) and the cast (credits.csv) for the title.

This dataset contains +9k unique titles on Amazon Prime with 15 columns containing their information, including:

genres: A list of genres.
production_countries: A list of countries that produced the title.
seasons: Number of seasons if it's a SHOW.
imdb_id: The title ID on IMDB.
imdb_score: Score on IMDB.
imdb_votes: Votes on IMDB.
tmdb_popularity: Popularity on TMDB.
tmdb_score: Score on TMDB.
And over +124k credits of actors and directors on Amazon Prime titles with 5 columns containing their information:

person_ID: The person ID on JustWatch.
id: The title ID on JustWatch.
name: The actor or director's name.
character_name: The character name.
role: ACTOR or DIRECTOR.

name	Name of the actor or director	Text	Helps identify key cast members
character	Name of the character played (if applicable)	Text	Useful for understanding casting patterns
role	Role type ‚Äî either ACTOR or DIRECTOR	Categorical	Used to separate director and actor analyses
